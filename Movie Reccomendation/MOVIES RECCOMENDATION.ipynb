```python
#Import Dependencies
```

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Data Collection and Preprocessing
```

```python
movies_data = pd.read_csv("C:/Users/DELL-PC/Documents/LESSONS/ML/Movie Reccomendation/movies.csv")
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
seletected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
```

```python
for feature in seletected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
#Text Vectorization
```

```python
vectorizer = TfidfVectorizer()
Feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
vectorizer
```

```python
#CosineSimilarity
```

```python
Similarity = cosine_similarity(Feature_vectors)
```

```python
print(Similarity.shape)
```

```python
# Finding Similar Movies
```

```python
list_of_all_movie_titles = movies_data['title'].tolist()
```

```python
# User Input
```

```python
movie_name = input('Enter the name of the movie')
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_movie_titles)
```

```python
close_match =find_close_match[0]
```

```python
# Finding Recommendations
```

```python
index_of_the_movie =movies_data[movies_data.title == close_match]['index'].values[0]
```

```python
similarity_score = list(enumerate(Similarity[index_of_the_movie]))
```

```python
# Sorting Movies by Similarity Score
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```

```python
print('Movies suggested for you : \n')
i = 1
```

```python
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i<30):
        print(i, '.', title_from_index)
        i+=1
```
